```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rayleigh_quotient(n, tol):
  A = np.random.rand(n, n)
  A += A.T
  v = np.random.rand(n, 1)
  v /= np.linalg.norm(v)

  k = 1
  mus = []
  while True:
    mu_k = v.T @ A @ v
    mus.append(mu_k[0][0])
    res = abs(np.linalg.norm(A @ v - mu_k * v) / mu_k)
    print(f'k = {k}: lam = {mu_k}\tres = {res}')

    if res <= tol:
      break
    w = np.linalg.solve((A - mu_k * np.identity(n)), v)
    v = w / np.linalg.norm(w)
    k += 1
  return mus
```

```python
def check_cubic(n, tol):
  mus = rayleigh_quotient(n, tol)
  errors = np.abs(np.array(mus[:-1]) - mus[-1])

  lim = errors[1:] / (errors[:-1]**3)

  plt.plot(lim)
  plt.title(f'n = {n}')
  plt.xlabel('$j - 1$')
  plt.ylabel('$e_{j + 1} / e_j^3$')
  plt.show()
```

```python
check_cubic(10, 1e-12)
```

```python
check_cubic(50, 1e-12)
```
